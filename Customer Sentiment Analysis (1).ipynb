```python
# Importing Libraries
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
from bs4 import BeautifulSoup

# Initialize WebDriver
driver = webdriver.Chrome()

# Open the URL
url = "https://www.flipkart.com/apple-iphone-15-blue-128-gb/product-reviews/itmbf14ef54f645d?pid=MOBGTAGPAQNVFZZY&lid=LSTMOBGTAGPAQNVFZZYQRLPCQ&marketplace=FLIPKART"
driver.get(url)

time.sleep(2)

Names = []
Reviews = []
Dates = []
Cities = []

for i in range(1, 11):
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, "html.parser")

    name = soup.find_all("p", {"class": "_2NsDsF AwS1CA"})
    for n in name:
        Names.append(n.text)

    city = soup.find_all("p", {"class": "MztJPv"})
    for c in city:
        Cities.append(c.text)

    date = soup.find_all("p", {"class": "_2NsDsF"})
    for d in date:
        Dates.append(d.text)

    review = soup.find_all("div", {"class": "ZmyHeo"})
    for r in review:
        Reviews.append(r.text)

    time.sleep(3)

    try:
        # Find the next button
        next_button = driver.find_element(By.XPATH, """/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[6]""") #till page-06

        # Scroll into view
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        time.sleep(1)

        # Click the next button
        next_button.click()
    except Exception as e:
        print("Could not click next button:", e)
        break  # Stop the loop if pagination fails

    time.sleep(3)

driver.quit()
```

```python
from datetime import datetime, timedelta

def convert_relative_date(text):
    if re.match(relative_pattern, text):
        months_ago = int(text.split()[0])  # Extract number
        target_date = datetime.today() - timedelta(days=months_ago * 30)  # Approximate month length
        return target_date.strftime("%b, %Y")  # Convert to "Jan, 2024" format
    return text

converted_dates = [convert_relative_date(item) for item in dates_only]
print(converted_dates)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from textblob import TextBlob
import string

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

df["reviews"] = df["review"].astype(str).apply(lambda x: sent_tokenize(x))
df
```

```python
# Function to calculate polarity
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

# Create a DataFrame with reviews
df = pd.DataFrame({'Review': Reviews})

# Apply polarity function
df['Polarity'] = df['Review'].apply(get_polarity)

# Display results
print(df[['Review', 'Polarity']])
```
